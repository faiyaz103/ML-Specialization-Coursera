## Problem Statement:
### Training set:
|   House size (1000 sqft)  |   Price ($1k) |
|---------------------------|---------------|
|   2.125                   |   400         |
|   1.445                   |   335         |

Now Predict price for a house size 1200 ftsq.


We need numpy, pandas and matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Notation:
- x = input variables/features                  : House size

- y = output variables/targets                  : House price

- m = total number of training example

- $f_{w,b}(x^{(i)}) = wx^{(i)} + b$           : model

- ($x^{(i)}$, $y^{(i)}$) = i'th training example

Let's assumew=100, b=200 for this example.

Define training set

```python
x=np.array([2.125,1.445])
y=np.array([400,335])

# Show training set in dataframe
df = pd.DataFrame({
    'House size (1000 sqft)': x,
    'House price ($1k)': y
})
print(df)
```

Total number of training example

```python
m = len(x)
print(f"Number of training example m = {m}")
```

Print all training example

```python
for i in range(m):
    print(f"(x^{i+1}, y^{i+1}) = ({x[i]}, {y[i]})")
```

Plot training set

```python
plt.scatter(x, y, marker='X', c='r') # Marker type X, Marker color red
plt.title('Housing Prices')
plt.xlabel('House Size (1000 sqft)')
plt.ylabel('House Price ($1k)')
plt.show()
```

Define the model with assumed values

```python
w=100
b=200
print(f"fw,b(x) = {w}x + {b}")
```

Calculate price for 1200 sqft

```python
xnew=1.2
Ypredict = w*xnew + b
print(f"Price for {xnew} sqft house is ${Ypredict}k")
```

Let's see how much our assumed value differs from training data set.
Create a function to calculate each fw,b(x) for w and b.

```python
def compute_model_output(x,w,b):
    fwb=np.zeros(m) # creates array with m zeroes
    for i in range(m):
        fwb[i] = w*x[i] + b
        # fwb[i]=fwb[i]/1000
    
    return fwb
```

Plot trainig set and calculated fw,b

```python
fwb=compute_model_output(x,w,b)
plt.plot(x,fwb,c='b',label='Our Prediction')
plt.plot(x,y,c='r',label='Training set values')
plt.xlabel('House Size (1000 sqft)')
plt.ylabel('House Price ($1k)')
plt.legend()
plt.show()
```
